## Get familiar with imagery

```python
image1 = io.imread('Andasol_09051987.jpg')
image2 = io.imread('Andasol_09122013.jpg')
```

```python
fig_width = 20
fig_height = 8

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(fig_width, fig_height))
ax1.imshow(image1, cmap="gray") 
ax2.imshow(image2, cmap="gray") 
ax3.imshow(abs(image1 - image2), cmap="hot") # crudely calc difference
```
